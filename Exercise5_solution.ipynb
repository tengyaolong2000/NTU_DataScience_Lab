```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("/Users/tengyaolong/Desktop/NTU_DataScience_Lab/train1.csv")
X=pd.DataFrame(df['SalePrice'])
Y=pd.DataFrame(df['CentralAir'])
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, shuffle=True)
```

```python
dectree2 = DecisionTreeClassifier(max_depth = 2)  # create the decision tree object
dectree2.fit(x_train, y_train)

f = plt.figure(figsize=(12,12))
plot_tree(dectree2, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["No","Yes"])

# Predict Legendary corresponding to Total Test
test_pred2 = dectree2.predict(x_test)

plt.show()
```

```python
# Print the Classification Accuracy
print("Classification Accuracy \t:", dectree2.score(x_test, y_test))

# Plot the two-way Confusion Matrix
sb.heatmap(confusion_matrix(y_test, test_pred2), 
           annot = True, fmt=".0f", annot_kws={"size": 18})
plt.show()
```

```python
#True Positive Rate 0.927835, False Positive Rate 0
```
